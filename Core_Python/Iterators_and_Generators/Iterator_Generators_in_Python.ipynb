```python
class Even:
    def __init__(self,start,max_limit):
        if start % 2 == 0 :
            self.start = start
        else :
            self.start = start + 1
        self.max_limit = max_limit
    def __iter__(self):
        self.val = self.start - 2
        return self
    def __next__(self):
        if self.val <= self.max_limit :
            self.val += 2 
            return self.val
        else :
            raise StopIteration("Max limit has Exceeded")
        
```

```python
one = Even(1,100)
one = iter(one)
```

```python
for var in one:
    print(var)
```

```python
class Prime:
    def __init__(self,start):
        self.start = start
    def __iter__(self):
        self.max_limit = 1000
        self.counter = 0
        return self
    def prime(self):
            num = self.start
            sq = round(sqrt(num)+1)
            for var in range(2,sq):
                if num % var == 0 :
                    self.start += 1
                    return self.prime()
            else :
                self.start += 1
                return num
    def __next__(self):
        if self.counter < self.max_limit :
            self.counter += 1
            return self.prime()
        else :
            raise StopIteration
```

```python
from math import sqrt
x = Prime(1)
x = iter(x)
c = 0
for var in x :
    print(var,end=', ')
    c = c+1
print("Total numbers = ",c)
```

```python
def even(num,max_limit):
    c = 0
    while c <= max_limit :
        if num % 2 == 0 :
            yield num
            num += 1
            c = c + 1
        else :
            num += 1

    
```
